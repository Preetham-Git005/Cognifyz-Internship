```python
# Importing the pandas library for data manipulation and analysis
import pandas as pd

# Reading the 'Railway_info.csv' file into a DataFrame
df = pd.read_csv('/content/Railway_info.csv')

# Calculating the total number of unique trains using the 'Train_No' column
num_trains = df['Train_No'].nunique()
print("Total number of unique trains:", num_trains)

# Calculating the number of unique source stations using the 'Source_Station_Name' column
unique_sources = df['Source_Station_Name'].nunique()
print("Number of unique source stations:", unique_sources)

# Calculating the number of unique destination stations using the 'Destination_Station_Name' column
unique_destinations = df['Destination_Station_Name'].nunique()
print("Number of unique destination stations:", unique_destinations)

# Finding the most common (frequent) source station using mode()
common_source = df['Source_Station_Name'].mode()[0]
print("Most common source station:", common_source)

# Finding the most common (frequent) destination station
common_destination = df['Destination_Station_Name'].mode()[0]
print("Most common destination station:", common_destination)
```

```python
# Importing the pandas library for data handling and analysis
import pandas as pd

# Loading the dataset from the given CSV file into a DataFrame
df = pd.read_csv('/content/Railway_info.csv')

# Displaying the number of missing (NaN) values in each column before cleaning
print("Missing values in each column:")
print(df.isnull().sum())

# Dropping all rows that contain any missing values and storing the result in a new DataFrame
df_cleaned = df.dropna()

# Displaying the number of missing values in each column after cleaning
print("\nMissing values after cleaning:")
print(df_cleaned.isnull().sum())

# Standardizing the format of 'Source_Station_Name' column
df_cleaned['Source_Station_Name'] = df_cleaned['Source_Station_Name'].str.upper().str.strip()

# Standardizing the format of 'Destination_Station_Name' column in the same way
df_cleaned['Destination_Station_Name'] = df_cleaned['Destination_Station_Name'].str.upper().str.strip()

# Displaying the cleaned and standardized values for source and destination station names
print("\nPreview of cleaned data:")
print(df_cleaned[['Source_Station_Name', 'Destination_Station_Name']].head())
```
